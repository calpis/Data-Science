## Training a Deep Q-Learning Model with Convolutional Neural Networks

Flappybird is an automatic side scrolling game where the user controls a bird and maneuvers it through the moving environment.  There are random pipes which are setup and the user must navigate above/below/between them to avoid contact.  By default, the altitude of the bird decreases, and the user can tap on the screen to let it ‘flap,’ increasing its altitude for a brief moment.  Repeated taps will result in steady incline of the bird's location.  Hitting any of the pipes or the top/bottom of the game screen will result in termination of the trial, and score is calculated based on how far the bird was able to travel.

The end goal for this project is to create an optimal action-selection policy for the game flappybird.  In essence, we want an unsupervised machine learning algorithm to make decision given the state the game is in, based on experience. 
For that we use the action-value function denoted as:$$Q(s,a) = r + \gamma*max(Q(s',a'))$$
where:<br>
s: current state<br>
a: action to be taken<br>
r: current reward<br>
gamma: future reward weight<br>
s': state after taking action a in state s<br>
a': all possible actions in s'<br>


This notebook will focus on two parts: 
1. Emulating the game flappybird and cleaning the data
2. Training a neural network from the data to learn weights to make an optimal action-selection policy.

To train the agent in this notebook, it took roughly 4 hours on a mac book pro, or 2 hours on an AWS p2.xlarge instance.  According to this article, 'https://yanpanlau.github.io/2016/07/10/FlappyBird-Keras.html', super human skill level was achieved after more than one million frames, which would have taken 40 hours on AWS, or 80 hours on a macbook.  At 4 hours of training, the agent does a good job of detecting certain features of the environment (ground and the pipe) and will be discussed.  Further training was avoided due to resource restriction and the tremendous amount of data required to achieve a comparable result to the article.

## Functions/Classes

We define multiple classes and functions below for organization.

- update_reward
    - used to test different reward values for different scenarios to see how it will impact training.
- PolicyAgent
    - Class which contains the CNN and methods to train/predict for different states
- ReplayData
    - Class which holds onto past data to run experience memory relay.
- init_game
    - initialize the game state so the code can interact with flappybird
- test_policy
    - once the agent is trained, we can call this function to test the performance
- train_policy
    - trains the agent by taking batches from ReplayData object.

```python
def update_reward(value):
    return value
#     #return value
#     if value==0.0:
#         return 0.1
#     elif value==1.0:
#         #print 'score!!'
#         return 1.0
#     elif value==-5:
#         return -5.0
#     elif value==-4.0: #crossed pipe and hit the pipe at the same time
#         return -5.0
#     else:
#         print 'unknown reward', value
#         return value


class PolicyAgent(object):
    '''agent to learn and provide optimal actions during the game run.
    the output is the expected reward for taking that action in that state
    output:
        higher index 0 -> flap
        higher index 1 -> do nothing
        '''
    def __init__(self, input_num_rows, input_num_cols):
        ##define cnn structur
        model = Sequential()
        model.add(Conv2D(32, 8, 8, subsample=(4, 4), border_mode='same',input_shape=(input_num_rows, input_num_cols, 4)))  #80*80*4
        model.add(Activation('relu'))
        model.add(Conv2D(64, 4, 4, subsample=(2, 2), border_mode='same'))
        model.add(Activation('relu'))
        model.add(Conv2D(64, 3, 3, subsample=(1, 1), border_mode='same'))
        model.add(Activation('relu'))
        model.add(Flatten())
        model.add(Dense(512))
        model.add(Activation('relu'))
        model.add(Dense(2))
        adam = Adam(lr=1e-4)
        model.compile(loss='mse',optimizer=adam) 
        self._model = model
        
        self._training_counter = 0
    
    
    def fit(self, stacked_images, y_data, batch_size=32, epochs=1):
        if type(stacked_images) is StackedImages:
            x_data = stacked_images.get_stacked_images()
        else:
            x_data = stacked_images
        #x_data = x_data.reshape(x_data.shape[1], x_data.shape[2], x_data.shape[0], 1)
        y_data = np.array(y_data)
        self._model.fit(x_data, y_data, batch_size, epochs, verbose=0)
        self._training_counter += 1
        
        
    def train_on_batch(self, x_data, y_data):
        '''
        inputs should be a stacked array. Does batch training to decrease overall training time
        compared to individual calls to 'fit.'
        '''
        if len(x_data) != len(y_data):
            print 'unequal lengths in data'
            return None
        shape_x = x_data[0].shape
        shape_y = y_data[0].shape
        for i in range(len(x_data)):
            y_data[i] = y_data[i].reshape(1,shape_y[0])
        inputs = np.concatenate(x_data, axis=0)
        targets = np.concatenate(y_data, axis=0)
        self._model.train_on_batch(inputs, targets)
        self._training_counter += inputs.shape[0]
        
    def predict(self, stacked_images):
        '''
        Outputs the action index of the suggested action.
        '''
        if type(stacked_images) is StackedImages:
            x_data = stacked_images.get_stacked_images()
            output = self._model.predict(x_data)
        else:
            output = self._model.predict(stacked_images)
        index = np.random.choice(np.flatnonzero(output==np.max(output)))

        return index, output[0]
    
    def save_weights(self, file_name):
        self._model.save_weights(file_name, overwrite=True)
         
    def summary(self):
        return self._model.summary()
    
    def load_weights(self, file_name):
        self._model.load_weights(file_name)
        self._model.compile(loss='mse', optimizer='sgd')

        
class ReplayData(object):
    '''    
    class used to keep track of experience replay.  While initializing the object,
    we define a max_size such that any items appended beyond the size limit will be 
    pushed, and oldest replay data will be popped. 
    
    replay_data: queue of lists
                 [0]: StackedImage object (current state)
                 [1]: Action
                 [2]: Reward
                 [3]: ImageProcessor object (next state)
                 [4]: is game terminal (boolean)'''
    def __init__(self, max_size=50):
        self._replay_data = deque()
        self._max_size = max_size
        
    @property 
    def max_size(self):
        return self._max_size
    
    @max_size.setter
    def max_size(self, value):
        self._max_size = value
    
    
    def append_smart(self, data, debug=False):
        self._replay_data.append(data)
        if debug:
            print 'debugging append_smart', self._max_size
        if len(self._replay_data) > self._max_size:
            self._replay_data.popleft()
        
    def size(self):
        return len(self._replay_data)
    
    def pop_left(self):
        return self._replay_data.popleft()
    
    def get_data(self):
        return self._replay_data
    
    def sample(self, count):
        return random.sample(self._replay_data, int(count))
        #self._replay_data = random.sample(self._replay_data, count)
        #return self
    
    def take_terminal_samples(self):
        return [i for i in self._replay_data if i[4]]
        #self._replay_data = [i for i in self._replay_data if i[4]]
        #return self
    
    def __getitem__(self, key):
        return self._replay_data[key]
    
    def __len__(self):
        return len(self._replay_data)
```

```python
def init_game(display_screen=False):
    '''initialize and return the game state'''
    game_state = game.GameState()
    s, reward, is_terminal = game_state.frame_step([1,0])
    return game_state

def test_agent(agent, debug=False, display_screen=False, display_image=False, plot=False, direction=None):
    '''
    tests the trained agent.  Depending on the arguments passed, it can output 
    various information that should help in debugging.
    '''
    game_state = init_game(display_screen)
    max_frames = 100
    total_reward = 0
    
    observations = deque()
    
    
    all_stacked_images = []
    
    a_initial = np.zeros([2])
    a_initial[0] = 1
    x_t, reward, is_terminal = game_state.frame_step(a_initial)

    x_t = skimage.color.rgb2gray(x_t)
    x_t = skimage.transform.resize(x_t,(input_num_rows,input_num_cols))
    x_t = skimage.exposure.rescale_intensity(x_t, out_range=(0,255))
    state_t = np.stack((x_t,x_t,x_t,x_t), axis=2)
    state_t = state_t.reshape(1, state_t.shape[0], state_t.shape[1], state_t.shape[2])

    
    for i in range(max_frames):
        #get suggested action
        action_t = np.zeros([2])
        if plot:
            plot_stack(state_t)
        action_index, output = agent.predict(state_t)
        action_t[action_index] = 1

        if debug:
            print action_index, output
        

        #get the state after action
        print action_t
        x_t1_c, reward_t, is_terminal = game_state.frame_step(action_t)
        
        if display_image:
            io.imshow(x_t1_c)
            plt.show()
        #reward = update_reward(.act(action))

        #new_state = p.getScreenRGB()
        x_t1 = skimage.color.rgb2gray(x_t1_c)
        x_t1 = skimage.transform.resize(x_t1,(input_num_rows,input_num_cols))
        x_t1 = skimage.exposure.rescale_intensity(x_t1, out_range=(0,255))
        x_t1 = x_t1.reshape(1, x_t1.shape[0], x_t1.shape[1], 1)

        #drop one and add new state to the experience replay element
        state_t1 = np.append(x_t1, state_t[:,:,:,:3], axis=3)

        #store in replay queue. 
        #replay_data.append_smart((state_t, action, reward, state_t1, is_terminal))

        total_reward += reward_t
        
        state_t = state_t1
        
        all_stacked_images.append(x_t1_c)

        if is_terminal:
            print 'reached terminal state'
            break
    return total_reward, all_stacked_images

    
def train_policy(agent, replay_data, take_terminal=False, batch_count=0, print_loss=False, epochs=1):
    '''use the replay data to train the network
    agent: PolicyAgent object
    replay_data: queue of lists
                 [0]: StackedImage object (current state)
                 [1]: Action
                 [2]: Reward
                 [3]: ImageProcessor object (next state)
                 [4]: is game terminal (boolean)
    take_terminal: train on only terminal samples
    batch_count: non-zero value indicates to use experience replay. 
                 # of frames to use from this replay_data to train the network
    '''

    if take_terminal:
        data_input = replay_data.take_terminal_samples()
        #data_input[0][0].plot()
    
    else:
        if batch_count!= 0:
            if batch_count < replay_data.size():
                data_input = replay_data.sample(batch_count)
    
    
    inputs = []
    targets = []
    
    for i in range(len(data_input)):
        curr_state = data_input[i][0]#.get_stacked_images()
        action = data_input[i][1] ##action index
        reward = data_input[i][2]
        is_terminal = data_input[i][4]
        next_state = data_input[i][3]

        #get suggested action and the reward for future state
        _, output = agent.predict(next_state)
        next_reward = np.max(output)
      
        
        if is_terminal:
            total_reward = reward
        else: #update optimal action-selection policy with expected reward
            total_reward = reward + gamma * next_reward

        output[action_index] = total_reward
        
        inputs.append(curr_state)
        targets.append(output)
        
    loss = agent.train_on_batch(inputs, targets)

    if print_loss:
        print 'loss is', loss, 'and agent_loss is', agent.get_loss()
        
    return agent
        
```

## Training

The below cell executes the training for the agent.  We define the maximum number of frames to explore, and define epsilon.  A key concept in reinforcement learning is the idea of exploration vs exploitation.  Exploration is the concept of choosing a random action to see what new states the agent can be exposed to, where as exploitation is the act of letting the agent choose the action to take.  If we rely 100% on exploitation, then the agent will never learn the rewards for certain uncharted scenarios, hence the ratio of exploration and exploitation is a key factor in reinforcement learning.  We utilize a method called *epsilon annealing* where we decrease epsilon gradually as the training time goes on.  

The epsilon annealing methodology recommends that we start off our epsilon at 1.0, which means we explore 100% of the time, not relying on agent prediction.  Then as we observe more scenarios and train the agent, we decrease epsilon until a very small number (in our case, we used 0.0001, which means we explore 0.01% of the time).  Intitively, as the model becomes more trained, we trust its decisions more.

The below took 50000 frames to explore, and the whole process took about 4 hours on a mac book pro.  Even on AWS, it still took 2 hours.

```python
#main program

#set up log files and clear content
output_file = 'logs/log'
reward_file = 'logs/rewards'
open(output_file, 'w').close()
open(reward_file, 'w').close()

scores = []
scored_states = []

prev_model_weights = 'model_weights'

read_previous_model = False

game_state = init_game()
max_frames = 50000 #total number of frames to play
exp_replay_size = 50000 #maximum size of the experience replay queue
num_to_observe = 3200 #number of frames to observe before we start training
batch_count = 32 #defines how many frames we consider as one batch to train the network

save_every = int(max_frames * 0.2) ##save the weights of the model every x frames
frames_per_action = 1 #only act on every x frames

#start timer to keep time log
start_time = timeit.default_timer()

count_game_over = 0

print 'training with batch_size:', batch_count

init_epsilon = 1.0
final_epsilon = 0.0001
epsilon = init_epsilon


observations = deque()
replay_data = ReplayData(max_size=exp_replay_size)
agent = PolicyAgent(input_num_rows, input_num_cols)

if read_previous_model:  #train further with pretrained agent
    agent.load_weights(prev_model_weights)


total_reward = 0


#get inital state state and make stacked object
a_initial = np.zeros([2]) #action array
a_initial[0] = 1

x_t, reward, is_terminal = game_state.frame_step(a_initial)
x_t = skimage.color.rgb2gray(x_t)
x_t = skimage.transform.resize(x_t,(input_num_rows,input_num_cols))
x_t = skimage.exposure.rescale_intensity(x_t, out_range=(0,255))
state_t = np.stack((x_t,x_t,x_t,x_t), axis=2)
state_t = state_t.reshape(1, state_t.shape[0], state_t.shape[1], state_t.shape[2])


for i in range(max_frames):
    '''
    two phases exists, observing and training.
    up to num_to_observe frames, we push all data into experience replay queue,
    then once we have enough frames in the queue, we use batch training.
    '''
    action_t = np.zeros([2])
    
    if i > num_to_observe and epsilon > final_epsilon: #decrease epsilon after finish observing
        epsilon -= (init_epsilon - final_epsilon)/max_frames 
        
        
    #log how long each x frames took
    if i % (max_frames/10) == 0:
        elapsed = timeit.default_timer() - start_time
        with open(output_file, 'a') as log_handle:
            log_handle.write('done with ' + str(i) + ' frames with ' + str(elapsed) + ' seconds\n')
        start_time = timeit.default_timer() 
        
        
    #save model weights every x iterations
    if i % save_every == 0 and i != 0:
        f_name = 'model_weights_'+str(i)
        agent.save_weights(f_name)

    #choose action
    if random.random() < epsilon: #random action
        if random.random() < 0.95:
            action_index = 0
        else:
            action_index=1
        action_t[action_index] = 1
    else: #suggested action
        action_index, output = agent.predict(state_t)
        action_t[action_index] = 1
        
    #get the state after action
    x_t1_c, reward_t, is_terminal = game_state.frame_step(action_t)
    
    x_t1 = skimage.color.rgb2gray(x_t1_c)
    x_t1 = skimage.transform.resize(x_t1,(input_num_rows,input_num_cols))
    x_t1 = skimage.exposure.rescale_intensity(x_t1, out_range=(0,255))
    x_t1 = x_t1.reshape(1, x_t1.shape[0], x_t1.shape[1], 1)
    
    #drop one and add new state to the experience replay element
    state_t1 = np.append(x_t1, state_t[:,:,:,:3], axis=3)
    
    #store in replay queue. 
    replay_data.append_smart((state_t, action_index, reward, state_t1, is_terminal))
    
    total_reward += reward
    
    #train once we have enough data
    if i > num_to_observe:
        train_policy(agent, replay_data, batch_count=batch_count)
        
    state_t = state_t1 
    
    if is_terminal:
        total_reward = 0
        game_state = init_game()

print 'finished'
```

3. After clearing the pipe, it starts moving up again due to lack of training in that scenario and crashes to the end tip of the top pipe.  What this experience would have yielded is for the agent to predict a different action for that scenario.
