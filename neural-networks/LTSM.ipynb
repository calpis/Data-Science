```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt



input_location = '/users/momori/data/sales-of-shampoo-over-a-three-ye.csv'
```

```python
def parser(x):
	return pd.datetime.strptime('190'+x, '%Y-%m')
series = pd.read_csv(input_location, header=0, parse_dates=[0], \
                     index_col=0, squeeze=True, date_parser=parser)
# summarize first few rows
print(series.head())
# line plot
series.plot()
plt.show()
```
