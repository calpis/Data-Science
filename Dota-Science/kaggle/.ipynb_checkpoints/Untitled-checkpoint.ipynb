```python
#organize dataframe such that each match_id is a unique row, and 
#the attributes are 'radiant_team', 'dire_team', and 'radiant_win'
# the '*_team' is a vector of size 112 (112 heroes total) with a value of 1 if that hero was selected
cols = ['match_id', 'hero_id', 'player_slot']
train_data = pd.merge(players[cols], matches.reset_index(), how='left')
train_data = train_data[['match_id','hero_id','radiant_win']]
train_data = train_data.head(50)
```

```python
train_data.groupby('radiant_win').size()
```

```python
def unstack_simplify(df):
    return df.unstack().iloc[10:].iloc[:11].reset_index(drop=True)

def make_radiant(df):
    vec = [0]*112
    rad_cols = [0,1,2,3,4]
    return df[rad_cols]

def make_dire(df):
    vec = [0]*112
    dire_cols = [5,6,7,8,9]
    return df[dire_cols]
    
#make binary vector for hero composition
def make_vec(df):
    print df.apply(make_dire)
```

```python
train_data_comp = train_data.groupby('match_id').apply(unstack_simplify)
make_vec(train_data_comp)
```
